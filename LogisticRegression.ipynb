# Feature selection with Recursive Feature Elimination (RFE) 
RFE feature selection trains a model on a subset of the features and iteratively removes the least important features one by one until we are left with the desired number of features.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# Load the dataset
X = df_supervised.drop(['hit', 'game_id','play_id','year', 'game_date','at_bat_number','bat_side','pitch_hand', 'count_balls_strikes', 'pitch_result'], axis=1)
y = df_supervised['hit']

# Create a logistic regression model
model = LogisticRegression(solver='lbfgs', max_iter=100)

# Use RFE to select the top 10 features
rfe = RFE(model, n_features_to_select=10)
rfe.fit(X, y)

# Get the selected feature indices
selected_indices = rfe.support_

# Get the names of all features
feature_names = X.columns  # Provide the names of features here

# Get the names of selected features
selected_feature_names = [feature_names[i] for i, selected in enumerate(selected_indices) if selected]

# Print the selected feature names
print("Selected Feature Names:", selected_feature_names)
```

# Feature selection with decision tree

```python
# Split data using 80-20 rule
# Dataset 1 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
# Create a Decision Tree Classifier note that the maximum number of leaf nodes is equal to the number of input features
from sklearn.tree import DecisionTreeClassifier
model = tree.DecisionTreeClassifier(max_leaf_nodes=len(X.columns), random_state=42,criterion="gini")
```

```python
# 5. Fit the classifier on the training data
model.fit(X_train, y_train)
```

```python
# 6. Dataset 1: Create a predictions list for test data
predictions = model.predict(X_test)

print(predictions)
```

```python
# 7. Calculate Accuracy
# Dataset 1
from sklearn.metrics import accuracy_score, balanced_accuracy_score
print(f"Dataset 1 Decision Tree Accuracy: {accuracy_score(y_true = y_test, y_pred = predictions)}")
```
